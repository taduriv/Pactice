```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from wordcloud import WordCloud
import nltk

# Download necessary NLTK data files
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

# Combine the text columns
df['Combined_Text'] = df['Message_Title'] + ' ' + df['Message_Text']

# Text cleaning and lemmatization
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
    # Handle potential float values and empty strings
    if isinstance(text, float) or text == "":
        return " "  # Return a single space instead of an empty string

    tokens = word_tokenize(text.lower())
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token.isalpha() and token not in stop_words]
    return ' '.join(tokens)

df['Cleaned_Text'] = df['Combined_Text'].apply(clean_text)

# ... rest of your code ...

# Sentiment analysis
analyzer = SentimentIntensityAnalyzer()

def get_sentiment_scores(text):
    return analyzer.polarity_scores(text)

df['Sentiment'] = df['Cleaned_Text'].apply(get_sentiment_scores)

# ... rest of your code ...

# Extract sentiment scores into separate columns
df['Compound'] = df['Sentiment'].apply(lambda score_dict: score_dict['compound'])
df['Positive'] = df['Sentiment'].apply(lambda score_dict: score_dict['pos'])
df['Negative'] = df['Sentiment'].apply(lambda score_dict: score_dict['neg'])
df['Neutral'] = df['Sentiment'].apply(lambda score_dict: score_dict['neu'])

# Classify sentiment
def classify_sentiment(compound_score):
    if compound_score >= 0.05:
        return 'Positive'
    elif compound_score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

df['Sentiment_Label'] = df['Compound'].apply(classify_sentiment)

# Create the final table
final_table = df[['Combined_Text', 'Sentiment_Label']]
print(final_table)

# Plot word count
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['Cleaned_Text'])

word_counts = X.sum(axis=0).A1
word_freq = pd.DataFrame({'Word': vectorizer.get_feature_names_out(), 'Count': word_counts})
word_freq = word_freq.sort_values(by='Count', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Word', data=word_freq.head(20))
plt.title('Top 20 Words by Count')
plt.xlabel('Count')
plt.ylabel('Word')
plt.show()

# Plot sentiment counts
plt.figure(figsize=(10, 6))
sns.countplot(x='Sentiment_Label', data=df)
plt.title('Sentiment Counts')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()

# Plot distribution of sentiment scores
plt.figure(figsize=(10, 6))
sns.histplot(df['Compound'], kde=True, bins=30)
plt.title('Compound Sentiment Score Distribution')
plt.xlabel('Compound Score')
plt.ylabel('Frequency')
plt.show()

# Boxplot of sentiment scores
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Positive', 'Negative', 'Neutral']])
plt.title('Boxplot of Sentiment Scores')
plt.xlabel('Sentiment')
plt.ylabel('Score')
plt.show()

# Generate and plot word cloud
all_words = ' '.join(df['Cleaned_Text'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_words)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Cleaned Text')
plt.show()

# Get positive and negative words from VADER lexicon
positive_words = {word for word, score in analyzer.lexicon.items() if score > 0}
negative_words = {word for word, score in analyzer.lexicon.items() if score < 0}

# Count positive and negative words
positive_word_counts = {}
negative_word_counts = {}

for word in vectorizer.get_feature_names_out():
    if word in positive_words:
        positive_word_counts[word] = word_counts[vectorizer.vocabulary_[word]]
    elif word in negative_words:
        negative_word_counts[word] = word_counts[vectorizer.vocabulary_[word]]

# Convert to DataFrame for plotting
positive_word_freq = pd.DataFrame.from_dict(positive_word_counts, orient='index', columns=['Count'])
positive_word_freq = positive_word_freq.sort_values(by='Count', ascending=False)

negative_word_freq = pd.DataFrame.from_dict(negative_word_counts, orient='index', columns=['Count'])
negative_word_freq = negative_word_freq.sort_values(by='Count', ascending=False)

# Plot positive word count
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y=positive_word_freq.index[:20], data=positive_word_freq.head(20)) # Use slicing to get the top 20 index labels
plt.title('Top 20 Positive Words by Count')
plt.xlabel('Count')
plt.ylabel('Positive Word')
plt.show()

# Plot negative word count
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y=negative_word_freq.index[:20], data=negative_word_freq.head(20)) # Use slicing for consistency
plt.title('Top 20 Negative Words by Count')
plt.xlabel('Count')
plt.ylabel('Negative Word')
plt.show()

# Additional plots for sentiment analysis

# Plot the distribution of positive, negative, and neutral scores
plt.figure(figsize=(10, 6))
sns.histplot(df[['Positive', 'Negative', 'Neutral']], kde=True, bins=30)
plt.title('Distribution of Positive, Negative, and Neutral Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()

# Boxplot of sentiment scores by sentiment label
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sentiment_Label', y='Compound', data=df)
plt.title('Boxplot of Compound Scores by Sentiment Label')
plt.xlabel('Sentiment Label')
plt.ylabel('Compound Score')
plt.show()

# Print the resulting dataframe
print(df[['Combined_Text', 'Cleaned_Text', 'Compound', 'Positive', 'Negative', 'Neutral', 'Sentiment_Label']])
```
